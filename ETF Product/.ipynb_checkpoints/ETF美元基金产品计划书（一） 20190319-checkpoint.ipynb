## ETF美元基金产品计划书

### 一、产品定位

满足国内高净值客户，在美元计价的资产配置需求，投资标的为市场主要ETF股票和债券指数。

通过预设规则（rule-base）策略和风险平价（risk parity）模型动态调整权重，平衡行业轮动的smart beta收益和重大危机事件回撤风险，为客户获取长期稳定的投资回报。

### 二、投资标的介绍

#### 1、The SPDR® S&P 500® ETF Trust

- The SPDR® S&P 500® ETF Trust（代码：SPY） 跟踪标普500指数（the S&P 500® Index）。

- 标普500指数包含500家美国最大的上市公司，覆盖全球行业分类标准（GICS）里定义的11个一级行业：

    - 基础材料(Materials)：化学品、金属采矿、纸产品和林产品；
    - 消费者非必需品(Consumer Discretionary)：汽车、服装、休闲和媒体；
    - 消费者常用品(Consumer Staples)：日用产品、食品和药品零售；
    - 能源(Energy)：能源设施、冶炼、石油和天然气的开采；
    - 金融(Financials)：银行、金融服务、保险和房地产；
    - 医疗保健(Healthcare)：经营型医疗保健服务、医疗产品、药品和生物技术；
    - 工业(Industrials)：资本货物、交通、建筑、航空和国防；
    - 技术(Technology)：硬件、软件和通讯设备；
    - 通信服务(Communication Services)：电信服务和无线通讯；
    - 公用事业(Utilities)：电力设备和天然气设备；
    - 地产业(Real Estate)：房地产开发、管理及相关信托。

- SPY于1993年1月发行，是美国第一支ETF基金，截至2019年3月，市值约2,628亿美元（https://us.spdrs.com/en/etf/spdr-sp-500-etf-SPY）。

```python
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
import seaborn as sns
import datetime
import pandas as pd
import numpy as np
import time
import requests
from itertools import combinations
from IPython.core.display import HTML
```

```python
# set end_date
end_date = '2018-12-31'
comparable_ticker = 'SPY'
```

```python
# read data from csv
etf_ticker = 'SPY'
csv_filename = etf_ticker+end_date+'.csv'
df = pd.read_csv(csv_filename)
```

```python
p_t=['Date','Adj Close']
p=df[p_t].sort_values(by='Date')
p['Date'] = pd.to_datetime(p['Date'])
p.set_index("Date", inplace=True)
p = p.rename({'Adj Close':etf_ticker}, axis=1)
spy = p
```

```python
# calc ma
mp = 200
spy['ma'] = spy[etf_ticker].rolling(mp,center=False,min_periods=1).mean()
```

```python
# Plot
row = 2
col = 2
gridsize = (row, col) # (row,col)
fig = plt.figure(figsize=(row*6, col*2*row))
ax1 = plt.subplot2grid(gridsize, (0, 0), colspan=col, rowspan=1)

ax1.plot(spy[etf_ticker], color='black', linewidth=2)
ax1.plot(spy['ma'], color='red', linewidth=1)
ax1.legend(loc='upper center')
```

#### 2、iShares 1-3 Year Treasury Bond ETF

The iShares 1-3 Year Treasury Bond ETF（代码：SHY）跟踪ICE U.S. Treasury 1-3 Year Bond Index指数，该指数由1-3年到期的美国政府发行的国库券（treasury bond）构成。

SHY于2002年7月发行，截至2019年3月，市值约202亿美元。SHY每月分红，按2019年2月分红计算的年化收益率约2.42%，按过去12个月累计分红计算的收益率为1.88%（https://www.ishares.com/us/products/239452/ishares-13-year-treasury-bond-etf#/）。

```python
# read data from csv
etf_ticker = 'SHY'
csv_filename = etf_ticker+end_date+'.csv'
df = pd.read_csv(csv_filename)
```

```python
p_t=['Date','Close','Dividends']
p=df[p_t].sort_values(by='Date')
p['Date'] = pd.to_datetime(p['Date'])
p.set_index("Date", inplace=True)
p = p.rename({'Close':etf_ticker}, axis=1)
```

```python
p['r']=p['Dividends']/p[etf_ticker]*12
```

```python
p=pd.merge(p,spy,left_index=True, right_index=True)
```

```python
# Plot
row = 2
col = 2
gridsize = (row, col) # (row,col)
fig = plt.figure(figsize=(row*6, col*4))
ax1 = plt.subplot2grid(gridsize, (0, 0), colspan=col, rowspan=1)
ax2 = plt.subplot2grid(gridsize, (1, 0), colspan=1, rowspan=1)
ax3 = plt.subplot2grid(gridsize, (1, 1), colspan=2, rowspan=1)

ax1.plot(p[etf_ticker], color='black', linewidth=2)
ax1x = ax1.twinx()
ax1x.plot(p[comparable_ticker],  linewidth=0.5)
ax1.legend(loc='upper center')
ax1x.legend(loc='lower center')

pm=p[p['r']>0]
r=pm['r'].values
step = (r.max()-r.min())/100.0
ax2.hist(r, bins=np.arange(r.min(), r.max()+step, step), align='left')
t='n:%d mean: %0.2f%% std: %0.2f%%' %(len(r), r.mean()*100,r.std()*100)
ax2.set_title(t)

ax3.axis('off')
```

### 三、投资策略

#### 1、筛选投资标的

- SPY与SPYG、SPYV、SPYD相关系数很高，成长、价值和红利因子对投资组合意义不大；
- 各行业中，除了XLE、XLF、XLC和XLRE，其他行业指数与SPY相关性较高，不具有另外配置的需要，XLE、XLF和XLRE与其他行业相关性低、且彼此之间相关性也不高，可以加入投资组合，XLC与SPY及其他行业均较低，但该ETF是从2018年开始发行，数据不够，保持观察但暂不加入投资组合；
- 各地区股票指数方面，SPDW、SPEM与SPY相关度较低，可以加入投资组合；
- 资产类别方面，债券固收指数SHY比SPAB与其他ETF相关对更低，因此将SHY加入投资组合。

```python
def read_csv(etf_ticker):
    csv_filename = etf_ticker[0]+end_date+'.csv'
    df = pd.read_csv(csv_filename)
    p_t=['Date','Adj Close']
    p=df[p_t].sort_values(by='Date')
    p['Date'] = pd.to_datetime(p['Date'])
    p.set_index("Date", inplace=True)
    p = p.rename({'Adj Close':etf_ticker[0]}, axis=1)
    pl=p

    for i in range(1,len(etf_ticker)):
        csv_filename = etf_ticker[i]+end_date+'.csv'
        df = pd.read_csv(csv_filename)    
        p=df[p_t].sort_values(by='Date')
        p['Date'] = pd.to_datetime(p['Date'])
        p.set_index("Date", inplace=True)
        p = p.rename({'Adj Close':etf_ticker[i]}, axis=1)
        pl = pd.merge(pl,p,left_index=True, right_index=True)
        
    return pl
```

```python
def plot_heatmap(corr, mirror=True):
    #Plot figsize
    fig, ax = plt.subplots(figsize=(len(etf_ticker)*0.8, len(etf_ticker)*0.8))

    #Generate Color Map
    colormap = sns.diverging_palette(220, 10, as_cmap=True)

    # Drop self-correlations
    if mirror == False:
        dropSelf = np.zeros_like(corr)
        dropSelf[np.triu_indices_from(dropSelf)] = True    
        #Generate Heat Map, allow annotations and place floats in map
        sns.heatmap(corr, cmap=colormap, annot=True, fmt=".2f", mask=dropSelf)
    else:
        sns.heatmap(corr, cmap=colormap, annot=True, fmt=".2f")

    #Apply xticks
    plt.xticks(range(len(corr.columns)), corr.columns, horizontalalignment='left')
    #Apply yticks
    plt.yticks(range(len(corr.columns)), corr.columns, verticalalignment='top')

    #show plot
    plt.show()
```

```python
# read data from csv
etf_ticker = ['SPY','SHY','SPYG','SPYV','SPYD','XLB','XLY','XLP','XLE','XLF','XLV','XLI','XLK','XLU','XLC','XLRE','SPDW','SPEM','SPAB']
pl=read_csv(etf_ticker)

mirror = True # False : show half
#Create Correlation
corr = pl.corr()
plot_heatmap(corr, mirror)
```
