```python
def get_price(code_name, code, start_date_str, end_date_str):
    p_t = ['trade_date', 'close'] #['trade_date', 'open', 'high', 'low', 'close', 'vol', 'pct_chg']
    
    #get data
    #price = pro.fund_daily(ts_code=code, start_date=start_date_str, end_date=end_date_str)
    price = pro.index_daily(ts_code=code, start_date=start_date_str, end_date=end_date_str)
    
    # set date index and resort data
    p = price[p_t].sort_values(by='trade_date')
    p['trade_date'] = pd.to_datetime(p['trade_date'])
    p.set_index("trade_date", inplace=True)
    
    p = p.rename({'close':code_name}, axis=1)
    
    return p
```

```python
def slope(pl, dw=5):
    sp = pl - pl
    x=np.arange(1,dw+1)
    for i in range(dw-1,len(pl)): 
        y=pl[i-dw+1:i+1].values / pl[i-dw+1] - 1
        #x = sm.add_constant((x))
        model = smf.OLS(y,x)
        results = model.fit()
        b = results.params
        sp[i] = b
    return sp
```

```python
def loss(pl,dw=5):
    sp = pl - pl
    for i in range(0,len(pl)-dw):
        sp[i]=min(pl[i+1:i+dw])/pl[i]-1
    return sp
```

```python
def ladder(pl, bw):
    sp=pl-pl
    bw1=bw[0:len(bw)-1]
    bw2=bw[1:len(bw)]
    for i in range(0,len(pl)):
        for j in range(0,len(bw1)):
            if bw1[j]<pl[i]<=bw2[j]:
                sp[i]=j
    ld=pl-pl-1
    for i in range(1,len(sp)):
        if sp[i]==0:
            ld[i]=1
        elif sp[i]==len(bw)-2:
            ld[i]=-1
        else:
            j=i-1
            z=0
            while j>=0 and z<=0:
                if sp[j]==sp[i]:
                    j=j-1
                else:
                    z=1
                    if sp[j]>sp[i]:
                        ld[i]=-1
                    else:
                        ld[i]=1
    return ld
```

```python
limit = 0.15
step = 0.05
bw=np.arange(-limit,limit+step,step)
bw=np.insert(bw, 0, -1)
bw=np.insert(bw, len(bw), 1)
```

### 以价格与均线的距离为x，与回报r做线性回归

```python
def liner_regression(x,y):
    model = smf.OLS(y,x)
    results = model.fit()
    b = results.params
    R = results.rsquared
    pvalue = results.pvalues
    t='Y=%0.4fX --- R2=%0.2f%% --- p-value=%0.4f' %(b[0], R*100, pvalue[0])
    return b,t
```
