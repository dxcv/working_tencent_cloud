```python
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
import datetime
import pandas as pd
import numpy as np
import time
import requests
from IPython.core.display import HTML
```

```python
import statsmodels.api as sm
import statsmodels.formula.api as smf
```

```python
# choose fund
start_date_str = '20150101'
end_date_str = '20181231'
```

```python
def get_price(code_name, code, start_date_str, end_date_str):
    p_t = ['trade_date', 'close'] #['trade_date', 'open', 'high', 'low', 'close', 'vol', 'pct_chg']
    
    #get data
    #price = pro.fund_daily(ts_code=code, start_date=start_date_str, end_date=end_date_str)
    price = pro.index_daily(ts_code=code, start_date=start_date_str, end_date=end_date_str)
    
    # set date index and resort data
    p = price[p_t].sort_values(by='trade_date')
    p['trade_date'] = pd.to_datetime(p['trade_date'])
    p.set_index("trade_date", inplace=True)
    
    p = p.rename({'close':code_name}, axis=1)
    
    return p
```

```python
# get data
ticks = {
    'ETF300':'399300.SZ' #'510310.SH'
}
c = 'ETF300'
p = get_price(c,ticks[c],start_date_str, end_date_str)
```

```python
# calc ma
mp = 90
p['ma'] = p[c].rolling(mp,center=False,min_periods=1).mean()
# calc diff
p['diff'] = p[c]/p['ma']-1

# return
p['r']=p[c].pct_change()

# slope
#p['d']=p['diff'].pct_change()
p['d']=p['diff'].diff()

p = p.drop(p.index[0:mp])
```

### 以价格与均线的距离为x，与回报r做线性回归

```python
x = p['diff'].values
y = p['r'].values
```

```python
# Regression
X=x*100
Y=y*100
model = smf.OLS(Y,X)
results = model.fit()
b = results.params
R = results.rsquared
pvalue = results.pvalues
t='Y=%0.4fX --- R2=%0.2f%% --- p-value=%0.4f' %(b[0], R*100, pvalue[0])
```

```python
# Plot
row = 2
col = 2
gridsize = (row, col) # (row,col)
fig = plt.figure(figsize=(row*6, col*4))
ax1 = plt.subplot2grid(gridsize, (0, 0), colspan=col, rowspan=1)
ax2 = plt.subplot2grid(gridsize, (1, 0), colspan=1, rowspan=1)
ax3 = plt.subplot2grid(gridsize, (1, 1), colspan=2, rowspan=1)

ax1.plot(p[c])
ax1.plot(p['ma'])

ax2.plot(X, Y, 'yo', X, X*b[0], '--k') 
ax2.set_title(t)

ax3.axis('off')
```

### 进一步细分

在前面的基础上，增加两组距离对回报/风险的影响

```python
p1=p[p['diff']<0]
p2=p[p['diff']>=0]
```

```python
fl = 0.15

p111=p1[p1['diff']<-fl]
p12=p1[p1['diff']>-fl]
p121=p12[p12['d']<0]
p122=p12[p12['d']>=0]

p211=p2[p2['diff']>fl]
p22=p2[p2['diff']<fl]
p221=p22[p22['d']>=0]
p222=p22[p22['d']<0]
```

```python
(len(p111)+len(p211))/len(p)
```

```python
def liner_regression(x,y):
    model = smf.OLS(y,x)
    results = model.fit()
    b = results.params
    R = results.rsquared
    pvalue = results.pvalues
    t='Y=%0.4fX --- R2=%0.2f%% --- p-value=%0.4f' %(b[0], R*100, pvalue[0])
    return b,t
```

```python
# Plot
row = 4
col = 2
gridsize = (row, col) # (row,col)
fig = plt.figure(figsize=(col*6, row*4))
ax1 = plt.subplot2grid(gridsize, (0, 0), colspan=col, rowspan=1)
ax2 = plt.subplot2grid(gridsize, (1, 0), colspan=1, rowspan=1)
ax3 = plt.subplot2grid(gridsize, (1, 1), colspan=1, rowspan=1)
ax4 = plt.subplot2grid(gridsize, (2, 0), colspan=1, rowspan=1)
ax5 = plt.subplot2grid(gridsize, (2, 1), colspan=1, rowspan=1)
ax6 = plt.subplot2grid(gridsize, (3, 0), colspan=1, rowspan=1)
ax7 = plt.subplot2grid(gridsize, (3, 1), colspan=1, rowspan=1)

ax1.plot(p[c],color='black', linewidth=2)
ax1.plot(p['ma'],color='red', linewidth=1)
#for vl in p11.index:
#    ax1.axvline(vl, ymin=0.0, ymax=1.0, alpha=0.5, color='#FE642E')
#for vl in p12.index:
#    ax1.axvline(vl, ymin=0.0, ymax=1.0, alpha=0.5, color='#58ACFA')

x = p111['diff'].values
y = p111['r'].values
b,t=liner_regression(x,y)
ax6.plot(x,y, 'yo', x, x*b[0], '--k') 
ax6.set_title(t)

x = p121['diff'].values
y = p121['r'].values
b,t=liner_regression(x,y)
ax2.plot(x,y, 'yo', x, x*b[0], '--k') 
ax2.set_title(t)

x = p122['diff'].values
y = p122['r'].values
b,t=liner_regression(x,y)
ax3.plot(x,y, 'yo', x, x*b[0], '--k') 
ax3.set_title(t)

x = p211['diff'].values
y = p211['r'].values
b,t=liner_regression(x,y)
ax7.plot(x,y, 'yo', x, x*b[0], '--k') 
ax7.set_title(t)

x = p221['diff'].values
y = p221['r'].values
b,t=liner_regression(x,y)
ax4.plot(x,y, 'yo', x, x*b[0], '--k') 
ax4.set_title(t)

x = p222['diff'].values
y = p222['r'].values
b,t=liner_regression(x,y)
ax5.plot(x,y, 'yo', x, x*b[0], '--k') 
ax5.set_title(t)
```
